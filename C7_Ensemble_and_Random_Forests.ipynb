```python
% matplotlib inline
import matplotlib
import matplotlib.pyplot as plt

import numpy as np
```

# Voting Classifier

```python
# moon data
from sklearn.datasets import make_moons
X, y = make_moons(n_samples=2000, noise=0.2)

def plot_dataset(X, y, axes):
    plt.plot(X[:, 0][y==0], X[:, 1][y==0], "bs")
    plt.plot(X[:, 0][y==1], X[:, 1][y==1], "g^")
    plt.axis(axes)
    plt.grid(True, which='both')
    plt.xlabel(r"$x_1$", fontsize=20)
    plt.ylabel(r"$x_2$", fontsize=20, rotation=0)

plot_dataset(X, y, [-1.5, 2.5, -1, 1.5])
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import VotingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC

log_clf = LogisticRegression(random_state=42)
rnd_clf = RandomForestClassifier(random_state=42)
svm_clf = SVC(probability=True, random_state=42) # 这里需要设置probability才能使用soft voting

voting_clf = VotingClassifier(
    estimators = [('lr', log_clf), ('rf', rnd_clf), ('svc', svm_clf)],
    voting='soft',
)
voting_clf.fit(X, y)
```

```python
from sklearn.metrics import accuracy_score
for clf in (log_clf, rnd_clf, svm_clf, voting_clf):
    clf.fit(X, y)
    y_pred = clf.predict(X)
    print(clf.__class__.__name__, accuracy_score(y, y_pred))
```

# Bagging and Pasting

```python
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier

bag_clf = BaggingClassifier(
    DecisionTreeClassifier(random_state=42), n_estimators = 500,
    max_samples = 1200, bootstrap=True, n_jobs = -1,
)
bag_clf.fit(X, y)
y_pred = bag_clf.predict(X)
print('Bagging predicts:', accuracy_score(y, y_pred))
```

### 关于Bagging & Pasting
1. Bagging会使Classification变得更平滑，也就是说less variance， more generalized.
2. bootstrap参数控制是Pasting(False)还是Bagging(True)，通常情况下Bagging会比Pasting有更高的bias，也就是更平滑

```python
# BaggingClassifier自检验
bag_clf = BaggingClassifier(
    DecisionTreeClassifier(random_state=42), n_estimators = 500,
    max_samples = 1200, bootstrap=True, n_jobs = -1,
    oob_score = True, # <<--
)
bag_clf.fit(X, y)
bag_clf.oob_score_
```

# Boosting

```python
# AdaBoost
# SAMME = Stagewise Additive Modeling using a Multiclass Exponential loss function
from sklearn.ensemble import AdaBoostClassifier
ada_clf = AdaBoostClassifier(
    DecisionTreeClassifier(max_depth=1), n_estimators=200,
    algorithm='SAMME.R', learning_rate=0.5,
)
ada_clf.fit(X, y)
y_pred = ada_clf.predict(X)
print("AdaBoost:", accuracy_score(y, y_pred))
```

```python
# GradientBoost
from sklearn.ensemble import GradientBoostingRegressor
gbrt = GradientBoostingRegressor(max_depth=2, n_estimators=3, learning_rate=1.0)
gbrt.fit(X, y)
```

#### AdaBoost
AdaBoost对feature设置权重，通过调整权重每轮训练一个越来越好的Classifier，

最终所有Classifier采用mode的方式进行predict。这里的算法还需要仔细再看一下

#### GradientBoost
GradientBoost通过递进式的训练，每次训练上次剩余的误差，最终预测时，

把所有Classifier的预测值进行相加，可以理解为对不同层次的细节进行预测，逐步平滑模型

```python
# Early stoping on Gradient Boosting
# 目的：找出最适合的Classifier数量（一个Classifier类似于一个细节层）
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

X_train, X_val, y_train, y_val = train_test_split(X, y, random_state=42)

# 一次性训练120个
gbrt = GradientBoostingRegressor(max_depth=2, n_estimators=800)
gbrt.fit(X_train, y_train)

errors = [mean_squared_error(y_val, y_pred) for y_pred in gbrt.staged_predict(X_val)]
bst_n_estimators = np.argmin(errors)
print('best estimators count:', bst_n_estimators)

gbrt_best = GradientBoostingRegressor(max_depth=2, n_estimators=bst_n_estimators)
gbrt_best.fit(X_train, y_train)
```

```python
# 在训练过程中提前退出，而不是先训练120个Classifier，再进行选择

# About warm_start
# When set to True, reuse the solution of the previous call to fit and add more 
# estimators to the ensemble, otherwise, just erase the previous solution.

gbrt = GradientBoostingRegressor(max_depth=2, warm_start=True)

min_val_error = float('inf')
error_going_up = 0
for n_estimators in range(1, 150):
    gbrt.n_estimators = n_estimators
    gbrt.fit(X_train, y_train) # 这里应该每次只会计算新增的estimator
    y_pred = gbrt.predict(X_val)
    val_error = mean_squared_error(y_pred, y_val)
    if val_error < min_val_error:
        min_val_error = val_error
    else:
        error_going_up += 1
        if error_going_up == 5:
            print('stop at n_estimators:', n_estimators - 5)
            break # early stopping
```
